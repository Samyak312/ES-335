Question - 1

How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.

Ans-1: We need to perform n x k x m multiplications and n x (k - 1) x m additions to multiply a (n, k) matrix with a (k, m) matrix. This is because each element of the product matrix is the dot product of a row of the first matrix and a column of the second matrix. A dot product of two vectors of length k requires k multiplications and k - 1 additions. Therefore, for each of the n rows of the first matrix and m columns of the second matrix, you need to perform k multiplications and k - 1 additions.

Questions - 2

Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
import numpy as np
import time

def matrix_mult_lists(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

def matrix_mult_numpy(A, B):
    return np.dot(A, B)

n, k, m = 50, 40, 30
A = [[np.random.randint(1, 10) for _ in range(k)] for _ in range(n)]
B = [[np.random.randint(1, 10) for _ in range(m)] for _ in range(k)]

start_time = time.time()
result_lists = matrix_mult_lists(A, B)
end_time = time.time()
lists_time = end_time - start_time

start_time = time.time()
result_numpy = matrix_mult_numpy(A, B)
end_time = time.time()
numpy_time = end_time - start_time

print("Time using lists:", lists_time)
print("Time using NumPy:", numpy_time)
print("NumPy is generally much faster due to vectorized operations and optimized algorithms for matrix operations.")
```

Question - 3

Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

```python
import numpy as np
import time

def median_sort(arr):
    sorted_arr = np.sort(arr)
    n = len(sorted_arr)
    if n % 2 == 0:
        return (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2
    else:
        return sorted_arr[n // 2]

arr = np.random.rand(10**6)

start_time = time.time()
result_sort = median_sort(arr)
print("Time using sort method:", time.time() - start_time)
print("Result using sort method:", result_sort)

start_time = time.time()
result_numpy = np.median(arr)
print("Time using numpy.median:", time.time() - start_time)
print("Result using numpy.median:", result_numpy)
```

Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.

y = x

y = x^2

y = sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100, 0.5)

# Plot 1
plt.figure()
plt.plot(x, x)
plt.title('y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 2
plt.figure()
plt.plot(x, x**2)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 3
plt.figure()
plt.plot(x, (x**3) / 100)
plt.title('y = (x^3)/100')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 4
plt.figure()
plt.plot(x, np.sin(x))
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 5
plt.figure()
plt.plot(x, np.sin(x) / x)
plt.title('y = sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 6
plt.figure()
plt.plot(x, np.log(x))
plt.title('y = log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot 7
plt.figure()
plt.plot(x, np.exp(x))
plt.title('y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Question - 10

Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as a, b, c, d, e. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1, 2, size=(20, 5))
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])
highest_std_column = df.std().idxmax()
lowest_mean_row = df.mean(axis=1).idxmin()

print(df)
print("Column with the highest standard deviation:", highest_std_column)
print("Row with the lowest mean:", lowest_mean_row)
```

Question - 12

Write a small piece of code to explain broadcasting in numpy.

```python
import numpy as np

# Example arrays
a = np.array([1, 2, 3])
b = 2

# Broadcasting: Multiply each element of array 'a' by scalar 'b'
result = a * b

print("Array 'a':", a)
print("Scalar 'b':", b)
print("Result after broadcasting:", result)
```

In this example, the scalar b is broadcasted to each element of the array a. The operation a * b is performed element-wise, and the result is an array where each element of a is multiplied by b. This is possible because NumPy automatically broadcasts the scalar b to match the shape of array a.

Broadcasting simplifies the syntax and makes operations more concise, improving code readability and efficiency.

Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np

def custom_argmin(arr):
    # Find the index of the minimum element using a loop
    min_val = arr[0]
    argmin_index = 0

    for i in range(1, arr.size):
        if arr[i] < min_val:
            min_val = arr[i]
            argmin_index = i

    return argmin_index

arr = np.array([4, 2, 7, 1, 9])
result = custom_argmin(arr)
print("Custom argmin:", result)

np_argmin_result = np.argmin(arr)
print("np.argmin:", np_argmin_result)
```
